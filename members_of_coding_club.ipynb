```python
'''
# Keith aka sl4ph34d
# coding club Aug 2020

# make a random set of coders with alphbetic names and assign them grades based on girl guides
# code must have running tests and use a dictionary
'''

import string
import numpy as np

'''
# class
'''

class Member:
    def __init__(self, name, grade):
        self.name   = name
        self.grade  = grade
        table       = (0,'rainbow'),(1,'brownie'),(2,'guide'),(3,'venture')
        grade_dict  = dict(table)
        self.rank   = grade_dict[self.grade]

'''
# getters
'''

def get_verbosity():
    verbosity   = '__notverbose__'
    return verbosity

'''
# makers
'''

# no test - too simple
def make_grades():
    grades = np.random.randint(0,4,size=20, dtype=int)
    return grades

def make_twenty_letters():
    letters     = list(string.ascii_lowercase)[:20]
    verbosity   = get_verbosity()
    if verbosity == '__verbose__':
        print('\n\n', letters)
    return letters

def make_letters_into_names(letters):
    name_parts = ['igel    ', 'ella   ', 'elmut   ', 'edolf   ', 'aswinder', 'acob    ', 'ina     ', 'obert   ']
    names = []
    for i in letters:
        name_index  = int(np.random.randint(0,8,size=1))
        this_name   = i + name_parts[name_index]
        names.append(this_name)
    verbosity   = get_verbosity()
    if verbosity == '__verbose__':
        print('\n\n', names)
    return names

def make_members(facts):
    names   = [i[0] for i in facts]
    grades  = [i[1] for i in facts]
    members = []
    for i in range(20):
        this_member = Member(names[i], grades[i])
        members.append(this_member)
    return members

'''
# tellers
'''

# no test - beyond edge
def tell_members(members):
    for member in members:
        print('Name: ', member.name,'\tRank: ', member.rank)

'''
#main
'''

if __name__ == '__main__':
    letters         = make_twenty_letters()
    names           = make_letters_into_names(letters)
    grades          = make_grades()
    facts           = list(zip(names, grades))
    
    verbosity   = get_verbosity()
    if verbosity == '__verbose__':
        print('\n\n', facts)
    
    members = make_members(facts)
    tell_members(members)
    
'''
# TESTS
'''

def test_make_letters_into_names():
    latters = make_twenty_letters()
    names   = make_letters_into_names(letters)
    assert isinstance(names, list), 'Error: should be a list'
    #print('pass')
    
def test_make_members():
    letters         = make_twenty_letters()
    names           = make_letters_into_names(letters)
    grades          = make_grades()
    facts           = list(zip(names, grades))
    members         = make_members(facts)
    assert isinstance(members, list),       'Error: should be a list'
    assert isinstance(members[0], Member),  'Error: should be a list'
    #print('pass')

def test_make_twenty_letters():
    letters = make_twenty_letters()
    assert len(letters) == 20, 'Error: should be twenty letters'
    #print('pass')

def test():
    test_make_members()
    test_make_twenty_letters()
    test_make_letters_into_names()
    print('\t\t\t\t\t\t\t\tcode passes unit tests')

test()
    
```
